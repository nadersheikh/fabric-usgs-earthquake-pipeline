Incremental Ingestion (Hourly)

Objective
Continuously ingest the latest USGS earthquake data into the Bronze layer for incremental processing.

Process Overview

Retrieve the latest hourly earthquake data from the USGS API.
Store the raw JSON file in the Bronze layer (/Files/bronze_usgs_hourly).
Preserve data in its original format without transformation to maintain raw data fidelity.

```python
import requests
from datetime import datetime
import notebookutils

# USGS hourly feed
url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_hour.geojson"

response = requests.get(url)
response.raise_for_status()

# Timestamp for file name
timestamp = datetime.utcnow().strftime("%Y%m%d_%H%M%S")

# Ensure folder exists (safe even if already exists)
notebookutils.fs.mkdirs("Files/bronze_usgs_hourly")

# Write file
file_path = f"Files/bronze_usgs_hourly/usgs_hourly_{timestamp}.json"
notebookutils.fs.put(file_path, response.text, True)

print(f"Done ingesting hourly file {file_path}")

notebookutils.notebook.exit(file_path)
```
